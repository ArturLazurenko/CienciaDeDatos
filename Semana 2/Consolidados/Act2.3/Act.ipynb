# INSERT

```python
from pymongo import MongoClient



cliente = MongoClient("mongodb://localhost:27017/")
db = cliente["Actividad2"]
coleccion = db["ACT2.3"]

empleados = [
    {"nombre": "Juan", "departamento": "IT", "salario": 50000},
    {"nombre": "Maria", "departamento": "HR", "salario": 45000},
    {"nombre": "Carlos", "departamento": "IT", "salario": 60000},
    {"nombre": "Ana", "departamento": "Finanzas", "salario": 55000},
    {"nombre": "Luis", "departamento": "IT", "salario": 52000},
    {"nombre": "Sofia", "departamento": "HR", "salario": 48000},
    {"nombre": "Pedro", "departamento": "Finanzas", "salario": 53000},
    {"nombre": "Laura", "departamento": "IT", "salario": 61000},
    {"nombre": "Diego", "departamento": "Ventas", "salario": 47000},
    {"nombre": "Elena", "departamento": "Ventas", "salario": 49000}
]

coleccion.insert_many(empleados)
print("Documentos insertados correctamente\n")
```

# READ

```python
print("Empleados del departamento IT:")
resultados = coleccion.find({"departamento": "IT"})

for emp in resultados:
    print(emp)
```

# Update

```python
print("\nActualizando salario de Juan...")
coleccion.update_one(
    {"nombre": "Juan"},
    {"$set": {"salario": 70000}}
)

empleado_actualizado = coleccion.find_one({"nombre": "Juan"})
print("Empleado actualizado:")
print(empleado_actualizado)
```

# DELETE

```python
print("\nEliminando empleado Diego...")
coleccion.delete_one({"nombre": "Diego"})

print("Empleados después de eliminar:")
for emp in coleccion.find():
    print(emp)
```
