# Ejercicio 3

```python
import json

data = {
    "empleados": [
        {"id": 1, "nombre": "Juan", "habilidades": ["Python", "SQL"]},
        {"id": 2, "nombre": "María", "habilidades": ["Java", "MongoDB"]},
        {"id": 3, "nombre": "Carlos", "habilidades": ["Python", "R"]}
    ]
}

# 1. Extraer los nombres
nombres = [emp["nombre"] for emp in data["empleados"]]
print("Nombres:", nombres)

# 2. Agregar una nueva habilidad a Juan
for emp in data["empleados"]:
    if emp["nombre"] == "Juan":
        emp["habilidades"].append("Docker")

# 3. Crear un nuevo empleado con id: 4
nuevo_empleado = {"id": 4, "nombre": "Ana", "habilidades": ["C++", "SQL"]}
data["empleados"].append(nuevo_empleado)

# 4. Eliminar las habilidades de María
for emp in data["empleados"]:
    if emp["nombre"] == "María":
        emp["habilidades"] = []

# Mostrar resultado final
print(json.dumps(data, indent=2, ensure_ascii=False))
```

# Ejercicio 4

```python
empleados = [
    {"id": 1, "nombre": "Juan", "salario": 50000},
    {"id": 2, "nombre": "María", "salario": 45000},
    {"id": 3, "nombre": "Carlos", "salario": 55000}
]

# 1. Agregar un nuevo empleado
empleados.append({"id": 4, "nombre": "Ana", "salario": 60000})

# 2. Buscar empleado por id
def buscar_empleado(id_buscar):
    for emp in empleados:
        if emp["id"] == id_buscar:
            return emp
    return None

print("Buscar ID 2:", buscar_empleado(2))

# 3. Calcular promedio de salarios
promedio = sum(emp["salario"] for emp in empleados) / len(empleados)
print("Promedio de salarios:", promedio)

# 4. Filtrar empleados con salario mas de 50000
altos_salarios = [emp for emp in empleados if emp["salario"] > 50000]
print("Salarios > 50000:", altos_salarios)

# 5. Actualizar el nombre del empleado con id=2
for emp in empleados:
    if emp["id"] == 2:
        emp["nombre"] = "Sofia"

# Mostrar lista final
print("Empleados actualizados:", empleados)
```
