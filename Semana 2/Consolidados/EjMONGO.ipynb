```python
# Insertar Datos en MongoDB
from pymongo import MongoClient

cliente = MongoClient("mongodb://localhost:27017/")
db = cliente["tienda"]
coleccion = db["productos"]


def insertar_datos_mongodb(datos):
    try:
        cliente = MongoClient('mongodb://localhost:27017/')
        db = cliente['clase_datos']
        coleccion = db['reportes']

        resultado = coleccion.insert_many(datos)
        print(f"Registros insertados")
    except Exception as e:
        print(f"Error: {e}")
    finally:
        cliente.close()
        print("Conexión cerrada")
insertar_datos_mongodb([
    {"nombre": "Laptop", "precio": 999, "categoria": "Electrónica"},
    {"nombre": "Mouse", "precio": 29, "categoria": "Electrónica"},
    {"nombre": "Escritorio", "precio": 299, "categoria": "Muebles"}
])
```

```python
from pymongo import MongoClient

cliente = MongoClient("mongodb://localhost:27017/")
db = cliente["tienda"]
coleccion = db["productos"]


for p in coleccion.find({"categoria": "Electrónica"}):
    print(p)


for p in coleccion.find({"precio": {"$lt": 100}}):
    print(p)


coleccion.update_one(
    {"nombre": "Laptop"},
    {"$mul": {"precio": 1.1}}
)


coleccion.delete_many({"precio": {"$lt": 50}})


coleccion.insert_one({
    "nombre": "Teclado",
    "precio": 49,
    "categoria": "Electrónica"
})

cliente.close()
```

# Ejercicio 6

```python
from pymongo import MongoClient

cliente = MongoClient("mongodb://localhost:27017/")
db = cliente["escuela"]
coleccion = db["estudiantes"]

# 1. Cursan Math
for e in coleccion.find({"materias": "Math"}):
    print(e)

# 2. Mayores de 20
for e in coleccion.find({"edad": {"$gt": 20}}):
    print(e)

# 3. Contar por edad
resultado = coleccion.aggregate([
    {"$group": {"_id": "$edad", "total": {"$sum": 1}}}
])

for r in resultado:
    print(r)

# 4. Solo nombres
for e in coleccion.find({}, {"nombre": 1, "_id": 0}):
    print(e)

cliente.close()
```
