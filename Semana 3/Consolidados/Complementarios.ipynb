## Ejercicio 3

```python
import math

# 1. 
def area_circulo(radio):
    return math.pi * radio**2

# 2. 
def celsius_a_fahrenheit(c):
    return (c * 9/5) + 32

# 3. 
def promedio(lista):
    return sum(lista) / len(lista)

# 4. 
def max_min(lista):
    return max(lista), min(lista)

print(area_circulo(5))
print(celsius_a_fahrenheit(30))
print(promedio([10, 20, 30]))
print(max_min([5, 2, 9, 1]))
```

# Ejercicio 5

```python
import numpy as np

v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

# 1. 
producto_punto = np.dot(v1, v2)

# 2. 
producto_cruz = np.cross(v1, v2)

# 3. 
mag_v1 = np.linalg.norm(v1)
mag_v2 = np.linalg.norm(v2)

# 4. 
norm_v1 = v1 / mag_v1
norm_v2 = v2 / mag_v2

# 
print("Producto punto:", producto_punto)
print("Producto cruz:", producto_cruz)
print("Magnitud v1:", mag_v1)
print("Magnitud v2:", mag_v2)
print("Normalizado v1:", norm_v1)
print("Normalizado v2:", norm_v2)
```

## Ejercicio 7

```python
import pandas as pd
import numpy as np

data = {
    'nombre': ['Ana', 'Luis', 'María', 'Carlos', 'Sofia'],
    'edad': [20, 22, 19, 21, 23],
    'carrera': ['Ing', 'Ing', 'Lic', 'Ing', 'Lic'],
    'promedio': [8.5, 9.0, 7.8, 8.2, 9.5]
}
df = pd.DataFrame(data)

# 1. 
df.loc[2, 'promedio'] = np.nan
df['promedio'] = df['promedio'].fillna(df['promedio'].mean())

# 2. 
df = pd.concat([df, df])
df = df.drop_duplicates()

# 3. 
df['edad_doble'] = df['edad'].apply(lambda x: x * 2)

# 4. 
loc_slice = df.loc[0:2, ['nombre', 'edad']]
iloc_slice = df.iloc[0:3, 0:2]

# 5. 
df_extra = pd.DataFrame({
    'nombre': ['Pedro'],
    'edad': [24],
    'carrera': ['Ing'],
    'promedio': [8.7]
})

df_concat = pd.concat([df, df_extra], ignore_index=True)

print(df)
print(loc_slice)
print(iloc_slice)
print(df_concat)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 
df = sns.load_dataset('iris')

# 
print(df.info())
print(df.head())

# 2.
print(df.describe())

# 3. 
df.hist()
plt.show()

# 4. 
corr = df.corr(numeric_only=True)
print(corr)

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.title("Matriz de correlación")
plt.show()

# 5. 
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title("Boxplot por especie")
plt.show()

# 6. 
Q1 = df['sepal_length'].quantile(0.25)
Q3 = df['sepal_length'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['sepal_length'] < Q1 - 1.5*IQR) | 
              (df['sepal_length'] > Q3 + 1.5*IQR)]

print("Outliers:")
print(outliers)
```

## Ejercicio 12

### ¿Qué es el ciclo CRISP-DM?
CRISP-DM es una metodología para desarrollar proyectos de ciencia de datos. Proporciona un enfoque estructurado para resolver problemas mediante datos, desde la comprensión del negocio hasta la implementación del modelo.

---

### Fases del proceso de ciencia de datos (CRISP-DM)

1. **Comprensión del negocio (Business Understanding)**    

2. **Comprensión de los datos (Data Understanding)**  

3. **Preparación de los datos (Data Preparation)** 

4. **Modelado (Modeling)** 

5. **Evaluación (Evaluation)**  

6. **Despliegue (Deployment)**  

---

## ¿Qué es el MVP (Minimum Viable Product) en ciencia de datos?

El **MVP (Producto Mínimo Viable)** es una versión inicial de un modelo o solución que:
- Tiene lo mínimo necesario para funcionar
- Permite validar rápidamente si la solución es útil
- Reduce tiempo y costo antes de desarrollar algo más complejo

En ciencia de datos, un MVP puede ser:
- Un modelo simple (ej. regresión lineal)
- Un dashboard básico
- Un análisis exploratorio con conclusiones iniciales
