# Ejercicio 3.1

```python
import pandas as pd

# Listas

numeros = [10, 20, 30, 40, 50]

# Diccionario
personas = [
    {"nombre": "Juan", "edad": 25, "salario": 30000},
    {"nombre": "Maria", "edad": 30, "salario": 50000},
    {"nombre": "Carlos", "edad": 22, "salario": 20000}
]

# DataFrame
df = pd.DataFrame(personas)

print(df)

# 2. LAMBDA

duplicar = lambda x: x * 2

print("Duplicar 5:", duplicar(5))


# 3. LIST COMPREHENSIONS

pares = [x for x in numeros if x % 20 == 0]

print("Números divisibles entre 20:", pares)


# 4. MANEJO DE ERRORES

try:
    numero = int(input("Ingresa un número: "))
    resultado = 10 / numero
    print("Resultado:", resultado)
except ZeroDivisionError:
    print("Error: No se puede dividir entre cero.")
except ValueError:
    print("Error: Debes ingresar un número válido.")
finally:
    print("Fin del bloque try-except.")

# 5. EJERCICIOS 

# Ejercicio 1: Suma de una lista
def suma_lista(lista):
    return sum(lista)

print("\nEjercicio 1:", suma_lista(numeros))


# Ejercicio 2: Número mayor
def numero_mayor(lista):
    return max(lista)

print("Ejercicio 2:", numero_mayor(numeros))


# Ejercicio 3: Contar elementos mayores a 25
def mayores_25(lista):
    return len([x for x in lista if x > 25])

print("Ejercicio 3:", mayores_25(numeros))


# Ejercicio 4: Convertir nombres a mayúsculas
def nombres_mayusculas(lista_personas):
    return [p["nombre"].upper() for p in lista_personas]

print("Ejercicio 4:", nombres_mayusculas(personas))


# Ejercicio 5: Promedio de salarios
def promedio_salarios(lista_personas):
    salarios = [p["salario"] for p in lista_personas]
    return sum(salarios) / len(salarios)

print("Ejercicio 5:", promedio_salarios(personas))
```

# Actividad 3.3

```python
import pandas as pd

# Cargar dataset 
ruta = "../DataSets/Titanic-Dataset.csv"
df = pd.read_csv(ruta)


# ESTADO INICIAL
print("=== ESTADO INICIAL ===")
print("Dimensiones:", df.shape)
print("\nValores nulos:")
print(df.isnull().sum())
print("\nDuplicados:", df.duplicated().sum())
display(df.head())

# Rellenar valores nulos

# Edad = promedio
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Embarque = moda
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Cabin = eliminar columna
if 'Cabin' in df.columns:
    df.drop(columns=['Cabin'], inplace=True)

# Eliminar duplicados
df = df.drop_duplicates()

# convertir a categoría
df['Survived'] = df['Survived'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')

# Estandarizar nombres de columnas
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")

# tamaño de familia
df['family_size'] = df['sibsp'] + df['parch'] + 1

# pasajero solo
df['is_alone'] = df['family_size'].apply(lambda x: 1 if x == 1 else 0)


# ESTADO FINAL
print("=== ESTADO FINAL ===")
print("Dimensiones:", df.shape)
print("\nValores nulos:")
print(df.isnull().sum())
print("\nDuplicados:", df.duplicated().sum())
display(df.head())
```
