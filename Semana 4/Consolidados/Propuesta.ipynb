```python
import pandas as pd
import numpy as np

# Crear un dataframe valores nulos
data = {
    'Nombre': ['Ana', 'Luis', 'Carlos', None],
    'Edad': [23, np.nan, 35, 29],
    'Ciudad': ['CDMX', 'Guadalajara', None, 'Monterrey']
}


df = pd.DataFrame(data)
display(df)
display(df.isnull())
display(df.notnull())

print(df.isnull().sum())

df.info()
```
