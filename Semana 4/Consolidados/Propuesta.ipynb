# Actividad 4.2

```python
# Crear un dataframe valores nulos
data = {
    'Nombre': ['Ana', 'Luis', 'Carlos', None],
    'Edad': [23, np.nan, 35, 29],
    'Ciudad': ['CDMX', 'Guadalajara', None, 'Monterrey']
}


df = pd.DataFrame(data)
display(df)

df_media = df.copy()
df_media['Edad'] = df_media['Edad'].fillna(df_media['Edad'].mean())

print("Imputación por media:")
display(df_media)

df_mediana = df.copy()
df_mediana['Edad'] = df_mediana['Edad'].fillna(df_mediana['Edad'].median())

print("Imputación por mediana:")
display(df_mediana)

df_moda = df.copy()

df_moda['Edad'] = df_moda['Edad'].fillna(df_moda['Edad'].mode()[0])
df_moda['Ciudad'] = df_moda['Ciudad'].fillna(df_moda['Ciudad'].mode()[0])
df_moda['Nombre'] = df_moda['Nombre'].fillna(df_moda['Nombre'].mode()[0])

print("Imputación por moda:")
display(df_moda)

df_ffill = df.ffill()
print("Forward fill:")
display(df_ffill)

df_bfill = df.bfill()
print("Backward fill:")
display(df_bfill)
```

## ¿Cuándo usar cada uno?
- Media
    - Cuando los datos son numéricos y simétricos

- Mediana
    - Cuando hay outliers (valores extremos)

- Moda
    - Para datos categóricos

- Forward / Backward Fill
    - Cuando hay orden temporal o secuencia


# Actividad 4.3

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

data = {
    'Nombre': ['Ana', 'Luis', 'Carlos', 'Sofía'],
    'Edad': [23, 45, 35, 29],
    'Salario': [10000, 20000, 15000, 12000],
    'Ciudad': ['CDMX', 'Guadalajara', 'Monterrey', 'CDMX']
}

df = pd.DataFrame(data)
display(df)

scaler = MinMaxScaler()

df_minmax = df.copy()
df_minmax[['Edad', 'Salario']] = scaler.fit_transform(df[['Edad', 'Salario']])

print("Normalización Min-Max:")
display(df_minmax)

scaler = StandardScaler()

df_std = df.copy()
df_std[['Edad', 'Salario']] = scaler.fit_transform(df[['Edad', 'Salario']])

print("Estandarización:")
display(df_std)

df_ohe = pd.get_dummies(df, columns=['Ciudad'])

print("One-Hot Encoding:")
display(df_ohe)

df_derivado = df.copy()

df_derivado['Salario_Anual'] = df_derivado['Salario'] * 12

print("Variables derivadas:")
display(df_derivado)
```

## Usos 
- MinMax
    - Datos en rango fijo
- ZScore
    - Datos con outliers (datos atipicos)
- One-Hot
    - Variables no numéricas
- Derivadas
    - Mejorar la información

# Actividad 4.4

```python
import pandas as pd
import numpy as np

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.feature_selection import SelectKBest, f_classif

# Carga de datos
df = pd.DataFrame({
    "edad": [25, 30, np.nan, 35],
    "salario": [5000, 6000, 5500, np.nan],
    "departamento": ["IT", "HR", np.nan, "Ventas"],
    "target": [1, 0, 1, 0]
})

X = df.drop("target", axis=1)
y = df["target"]

# Columnas
num_cols = ["edad", "salario"]
cat_cols = ["departamento"]

# Pipeline
pipeline = Pipeline([
    ("preprocessing", ColumnTransformer([
        ("num", Pipeline([
            ("imputer", SimpleImputer(strategy="mean")),
            ("scaler", StandardScaler())
        ]), num_cols),
        
        ("cat", Pipeline([
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("encoder", OneHotEncoder())
        ]), cat_cols)
    ])),
    
    ("feature_selection", SelectKBest(score_func=f_classif, k=2))
])

X_procesado = pipeline.fit_transform(X, y)

display(X_procesado)
```
